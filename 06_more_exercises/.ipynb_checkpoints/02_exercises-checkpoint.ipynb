#### Ex 1 - Solution of algebraic equations
Consider the points $A(0,0)$ and $B(2,0)$ in the plane. We look for a point $C(x,y)$ such that $\|BC\|=2$ and the area $\mathcal{A}$ of the triangle $T$ of vertices $(A,B,C)$ is equal to $2$.

Let us recall that the area of the triangle $T$ is given by 

$$
\mathcal{A} = \frac{1}{2}|\det(AB,AC)|.
$$

1. Set the problem under a system of algebraic equations to solve. Write the corresponding function $\mathbf{F}(\mathbf{x})$ for which the problem corresponds to the solution of $\mathbf{F}(\mathbf{x})=\mathbf{0}$.

2. By using ```scipy.optimize.fsolve()```, solve the problem and print the coordinates $(x,y)$ of the point $C$.

3. Bonus: draw the resulting triangle in the plane.

````python
import numpy as np

pA = [0, 0]
pB = [2, 0]

area_val = 2
xA, yA = pA
xB, yB = pB

def pointC(pC):
    xC, yC = pC
    det = 2*yC
    BC = (xC-2)**2 + yC**2

    f1 = det - 2*area_val
    f2 = BC - 4

    F = np.array([f1,f2])

    return F
````

````python
c0 = [1, 2]

from scipy.optimize import fsolve
pC = fsolve(pointC, c0)
xC, yC = pC

print("C = ", pC)

import matplotlib.pyplot as plt

plt.scatter(xA, yA, label = "A")
plt.scatter(xB, yB, label = "B")
plt.scatter(xC, yC, label = "C")

plt.plot([xA, xB], [yA, yB], lw=3)
plt.plot([xB, xC], [yB, yC], lw=3)
plt.plot([xC, xA], [yC, yA], lw=3)

plt.grid()
plt.legend()
````

````python
#3
def ffprim(x):
    f = np.arctan(x)
    fprim = 1 / (1 + x**2)
    return f, fprim

def newton_scalar(x0, ffprim, n):
    xn = np.zeros(n+1)
    xn[0] = x0
    for i in range(n):
        f, fprim = ffprim(xn[i])
        xn[i+1] = xn[i] - f / fprim
    n_values = np.arange(n+1)
    return xn, n_values

x0 = 1.0
xn, n = newton_scalar(x0, ffprim, 21)

plt.scatter(n, xn)
plt.xlabel("Iteration")
plt.ylabel("x_n")
plt.grid()
plt.show()
````

#### Ex 3 - Population dynamics: Lotka-Volterra system
Consider the so-called Lotka-Volterra (prey-predator) dynamical system

$$
\begin{align*}
& \frac{du}{dt} = (1-v)u, \\
& \frac{dv}{dt} = (u-1)v.
\end{align*}
$$

and the initial condition $u(0)=u_0$, $v(0)=v_0$. Let
$\mathbf{x}(t)=(u(t),v(t))^T$. This system can be written in the usual vector form $\dfrac{d\mathbf{x}(t)}{dt} = \mathbf{F}(\mathbf{x}(t),t)$.

1. Write a function

```
xdot = Lotka(x, t) 
```

that implements the function of the Lotka-Volterra dynamical system and returns $\mathbf{F}(\mathbf{x},t)$.

````python
import numpy as np


def Lotka(x,t):
    u, v = x
    dudt = (1-v)*u
    dvdt = (u-1)*v
    F = np.array([dudt, dvdt])
    return F
````

2. With the help of ```scipy.integrate.odeint()```, numerically solve the Lotka-Volterra differential problem with the initial data $u_0=4$ and $v_0=1$. One could use the following array of discrete times:

```
T=20
td = np.linspace(0, T, 200).
```

Plot the discrete solutions $(t_n, u_n)$, $(t_n,v_n)$ as time series and $(u_n,v_n)$ in the phase plane $(u,v)$, obtained with ```odeint()```.

````python
from scipy.integrate import odeint

T = 20
td = np.linspace(0, T, 200)
u0 = 4
v0 = 1
x0 = np.array([u0, v0])

xsol = odeint(Lotka, x0, td)

u = xsol[:,0]
v = xsol[:,1]

import matplotlib.pyplot as plt

plt.figure(figsize = (6,7))

plt.subplot(3,1,1)
plt.plot(td, u, 'r')
plt.title("u(t)")
plt.xlabel("t")
plt.ylabel("u")
plt.grid()

plt.subplot(3,1,2)
plt.plot(td, v, 'b')
plt.title("v(t)")
plt.xlabel("t")
plt.ylabel("v")
plt.grid()

plt.subplot(3,1,3)
plt.plot(u, v, 'g')
plt.title("Phase Plane (u,v)")
plt.xlabel("u")
plt.ylabel("v")
plt.grid()

plt.tight_layout()
````

3. For the Lotka-Volterra system, the following quantity is conserved during time:

$$
\eta(\mathbf{x}) = u - \text{ln}(u) + v - \text{ln}(v).
$$

Plot the discrete curve $(t_n,\eta_n)$, where 

$$
\eta_n = u_n - \text{ln}(u_n) + v_n - \text{ln}(v_n).
$$

Change the 'y' scale of the plot (`plt.ylim()`) for better interpretation of the result at correct bounds.

````python
def eta(x):
    u = x[:,0]
    v = x[:,1]
    return u - np.log(u) + v - np.log(v)

etasol = eta(xsol)

plt.figure(figsize=(6,4))
plt.plot(td, etasol, 'm', label=r'$\eta_n = u_n - \ln(u_n) + v_n - \ln(v_n)$')
plt.xlabel('t')
plt.ylabel(r'$\eta_n$')
plt.title('Invariant Quantity over Time')
plt.grid(True)

# Setting correct bounds for clearer view
plt.ylim(min(etasol) - 0.01, max(etasol) + 0.01)

plt.legend()
plt.tight_layout()
plt.show()
````

#### Ex 4 - Ralston's scheme with application to the Lotka-Volterra system
The Ralston's scheme is yet another predictor-corrector explicit numerical scheme for differential equations or systems of equations. It reads

$$
\begin{align*}
& k_1 = \mathbf{F}(\mathbf{x}_n, t_n), \\
& k_2 = \mathbf{F}(\mathbf{x}_n+\frac{2}{3}\Delta t\, k_1, t_n+\frac{2}{3}\Delta t), \\
& \mathbf{x}_{n+1} = \mathbf{x}_n 
+ \Delta t\, \left\{ \frac{1}{4}k_1 + \frac{3}{4}k_2 \right\}.
\end{align*}
$$

Write an `python` function

```
td, xsol = Ralston(F, x0, T, N)
```

that implements the Ralston's scheme  for a function $F(\mathbf{x},t)$, with an initial condition $x_0$ and a final time $T$, using a time discretization made of $N$ small intervals, with a constant time step $\Delta t=\frac{T}{N}$. 
Output arrays are the discrete time array `td` and the array  `xsol` of discrete vector states $\mathbf{x}_n$ at discrete times $t_n=n\Delta t$.
.

````python
def Ralston(F, x0, T, N):
    d = x0.size
    h = T / N;
    td = np.linspace(0, T, N+1)
    x = np.zeros((N+1, d))
    x[0,:] = x0
    for n in range(N):
        k1 = F(x[n, :], td[n])
        k2 = F(x[n, :] + 2/3*h*k1, td[n] + 2/3*h)
        x[n+1, :] = x[n, :] + h*(1/3*k1 + 3/4*k2)
    return td, x
````
